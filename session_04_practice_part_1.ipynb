### Handling Common Inconsistencies

**Text: Mixed cases and extra spaces**

```python
# Clean customer names
df['user_name'] = df['user_name'].str.strip().str.title()
```

**Units: Inconsistent measurements**

```python
def convert_weight(x):
    if not isinstance(x, str):
        return x

    if 'kg' in x:
        return float(x.replace('kg', ''))
    if 'lbs' in x:
        return float(x.replace('lbs', '')) * 0.453592

df['weight_kg'] = df['weight'].apply(convert_weight)
df.drop('weight', axis=1, inplace=True)
df
```

**Categories: Different spellings**

```python
df.country.unique()
```

```python
# Standardize country names
country_mapping = {
    'USA': 'United States',
    'U.S.A.': 'United States',
    'UK': 'United Kingdom',
    'canada': 'Canada'
}
df['country'] = df['country'].replace(country_mapping).str.title()
```

### Handling Outliers

**Detecting outliers using Interquartile Range (IQR)**

```python
Q1 = df.user_salary.quantile(0.25)
Q3 = df.user_salary.quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df["user_salary"] < (Q1 - 1.5 * IQR)) | (df["user_salary"] > (Q3 + 1.5 * IQR))]
outliers
```

```python
# Removing outliers
filtered_df = df[~df.index.isin(outliers.index)]
```

### Identifying the Highest-Paid Employees by Country


```python
top_salaries = filtered_df.groupby(df.country)["user_salary"].max()
print(top_salaries)
```

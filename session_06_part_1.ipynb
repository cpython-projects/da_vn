```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('real_estate_data.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.info()
```

```python
df['balcony'] = df.balcony.fillna(0)
df['balcony'] = df.balcony.astype('int64')
```

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'])
df['is_apartment'] = df.is_apartment.fillna('False')
df['is_apartment'] = df.is_apartment.astype('bool')
```

```python
df.info()
```

```python
fig = px.histogram(df, x='last_price', nbins=100, title='Distribution of Apartment Prices')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.box(df, x='rooms', y='last_price', title='Price Distribution by Room Count')
fig.show()
```

```python
import plotly.figure_factory as ff

price_data = df['last_price'].dropna()
fig = ff.create_distplot([price_data], group_labels=['Price'], show_hist=False)
fig.update_layout(title='Price Density Plot')
fig.show()
```

```python
fig = px.scatter(df, x='total_area', y='last_price', title='Price vs Total Area')
fig.show()
```

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'])
df_time = df.groupby(df['first_day_exposition'].dt.to_period('M'))['last_price'].mean().reset_index()
df_time['first_day_exposition'] = df_time['first_day_exposition'].astype(str)

fig = px.line(df_time, x='first_day_exposition', y='last_price', title='Average Price by Month')
fig.show()
```

```python
df_bar = df.groupby('locality_name')['last_price'].mean().nlargest(10).reset_index()

fig = px.bar(df_bar, x='locality_name', y='last_price', title='Top 10 Cities by Average Price')
fig.show()
```

```python
import pandas as pd
df = pd.read_csv('session_04_data (1).csv', delimiter=',', encoding='utf-8')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.columns.tolist()
```

```python
df.rename(columns={
    'ID': 'id',
    'Name': 'user_name',
    'gender': 'user_gender',
    'Age': 'user_age',
    'Salary ($)': 'user_salary',
    'Hire Date': 'user_hire_date',
    'Country': 'user_country',
    'Weight': 'user_weight'
}, inplace=True)
df.columns.tolist()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
x = df.duplicated()
df[x]
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
x = uniques = df.user_age.dropna().unique()
print(x)
```

```python
df['user_age'] = df.user_age.replace(
    {
        'Thirty': '30',
        'Twenty': '20'
    }
)
df['user_age'] = pd.to_numeric(df['user_age'])

age_median = df.user_age.median()
df['user_age'] = df.user_age.fillna(age_median)

df.user_age
```

```python
df.user_hire_date.dropna().unique()
```

```python
pd.to_datetime(df.user_hire_date, errors='coerce')
```

```python
from dateutil import parser

def date_parser(date_as_string):
  if pd.notna(date_as_string):
    return parser.parse(date_as_string).strftime('%Y-%m-%d') # %d.%m.%Y
  return date_as_string

df.user_hire_date = df.user_hire_date.apply(date_parser)
df.user_hire_date
```

```python
df['user_hire_date'] = pd.to_datetime(df['user_hire_date'])
df.user_hire_date
```

```python
df.user_gender = df.user_gender.astype('category')
df.user_gender
```

```python
def converter_weight(item):
  if not isinstance(item, str):
    return item
  item = item.lower()
  if 'kg' in item:
    return float(item.replace('kg', ''))
  if 'lbs' in item:
    return float(item.replace('lbs', '')) * 0.453592

df['user_weight_kg'] = df['user_weight'].apply(converter_weight)
df.drop('user_weight', axis=1, inplace=True)

df.user_weight_kg
```

```python
df
```

```python
df.user_name = df.user_name.str.strip().str.title()

df
```

```python
df.user_country.dropna().unique()
```

```python
country_mappings = {
    'USA': 'United States',
    'U.S.A.': 'United States',
    'UK': 'United Kingdom',
    'canada ': 'Canada'
}

df['user_country'] = df['user_country'].replace(country_mappings).str.title()

df
```

```python
df.user_salary = df.user_salary.fillna(df.user_salary.mean())
df
```

```python
df.info()
```

```python
q1 = df.user_salary.quantile(0.25)
q3 = df.user_salary.quantile(0.75)
iqr = q3-q1
outliers_salary = df[(df.user_salary < q1-1.5*iqr) | (df.user_salary > q3+1.5*iqr)]
outliers_salary
```

```python
filtering_data = df[~df.index.isin(outliers_salary)]
```

```python
filtering_data
```

```python
top = filtering_data.groupby(df.user_country)['user_salary'].max()
print(top)
```
